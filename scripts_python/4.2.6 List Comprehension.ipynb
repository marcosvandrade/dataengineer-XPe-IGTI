# List

```python
# função que transformfora temperatura de celsisus para Fahrenheit
def converte_celsisus_to_fahrenheit(T):
    return round((float(9)/5)*T + 32,2)
```

```python
tempertaturas = [27, 28.5, 32, 35, 18.2, 16.5,]
```

```python
#para cada elemento chamamos a função
[converte_celsisus_to_fahrenheit(temp) for temp in tempertaturas]
```

```python
#com condição
[converte_celsisus_to_fahrenheit(temp) for temp in tempertaturas if temp > 20]
```

# Desafio

Para todos os números de 1 a 1000, encontre maior número divisível para cada numero (1 a 10)

```python
numeros = list(range(1,1001))
```

```python
{num:max([divisor for divisor in range(1,10) if num % divisor == 0]) for num in numeros}
```
